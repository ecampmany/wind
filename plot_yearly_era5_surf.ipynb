```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vtx_url2='https://api.vortexfdc.com/data/386587/f39e_cLEGF2T9vbjqB7Ruda6_aaU2uVQ/vortex.reanalysis.Met%20mast%20609.ERA5.txt'
era5=pd.read_csv(vtx_url2, skiprows=4, sep='\s+', names=['date','time', 'speed', 'dir', 'm100','d100','T','P'], parse_dates=[['date','time']])
```

```python
print(era5)
fig, axs = plt.subplots(figsize=(12, 4))
era5.groupby(era5["date_time"].dt.year)["m100"].mean().plot(rot=0, ax=axs, label="ERA5 100m")
plt.xlabel("Year");  
plt.ylabel("Wind Speed $(m/s)$");
plt.title("Yearly Mean Averages RAW Reanalysis")
plt.legend()
```
